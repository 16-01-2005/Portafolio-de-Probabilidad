# **Clasificación de Variables**
## ***Variables relevantes y su clasificación***

## ***Variables Discretas***
### **1. Subnational1 (Provincia o subdivisión regional)**

La variable Subnational1 es de naturaleza discreta y se clasifica específicamente como una variable categórica nominal. La variable representa divisiones administrativas o provincias dentro del territorio analizado, como por ejemplo Loja, Azuay o Zamora Chinchipe.

### **2. Threshold (Umbral)**

Esta variable es de naturaleza discreta y, específicamente, se clasifica como ordinal. En el contexto de los datos forestales, representa niveles porcentuales predefinidos de cobertura del dosel arbóreo (por ejemplo: 10%, 15%, 20%, 25%, 30%, 50% y 75%).

No es una escala infinita, sino un conjunto de valores finitos seleccionados para actuar como filtros. Su propósito es definir el criterio técnico de lo que se considera "bosque" para una región específica antes de realizar los cálculos de área.

---

## ***Variables Continuas***

### **3.	Area_ha (Área total)**

La variable Area_ha es de naturaleza continua, ya que representa una medición cuantitativa de la superficie física total de una unidad geográfica expresada en hectáreas.
Este tipo de variable puede tomar cualquier valor dentro de un intervalo, incluyendo valores decimales, dependiendo de la precisión de la medición.

Se clasifica como continua porque proviene de un proceso de medición física del espacio geográfico y no de un conteo de elementos discretos.


### **4. Extent_2000_ha (Extensión en el año 2000)**

La variable es continua, ya que representa la superficie de cobertura forestal en el año 2000 medida en hectáreas. Esta variable surge de estimaciones espaciales derivadas de imágenes satelitales y modelos de cobertura terrestre.

Al ser una magnitud física, puede tomar valores decimales y cualquier valor dentro de un rango continuo, dependiendo de la resolución del análisis geoespacial.


### **5. Tc_loss_ha_2024 (Pérdida de cobertura arbórea)**

Esta variable es continua, ya que representa la cantidad de superficie perdida de cobertura arbórea durante el año 2024, expresada en hectáreas.

Esta variable proviene de mediciones espaciales derivadas de observaciones satelitales, por lo que puede tomar valores decimales y cualquier valor dentro de un intervalo continuo.


## ***Espacio muestral de cada variable***

#### 1. Variables Discretas (Espacio muestral finito y enumerable): Para estas variables, el espacio muestral es un conjunto exacto y contable de opciones
---

El espacio muestral exacto es: {0%, 10%, 15%, 20%, 25%, 30%, 50%, 75%}

#### 2. En su lugar, el universo de valores que pueden tomar se define mediante un intervalo continuo conocido como rango teórico o lógico

* **Area_ha** *(Área total)*: Su espacio de valores válidos no puede ser negativo y está limitado por la realidad geográfica; va desde un valor mayor a 0 hasta aproximadamente 4,000,000 de hectáreas, que corresponde a la extensión de las provincias más grandes del país.

* **Extent_2000_ha** *(Extensión en el año 2000)*: Su espacio de valores continuos va desde 0 hasta el límite exacto dictado por Area_ha. Físicamente, la cobertura forestal inicial de una provincia nunca puede superar el tamaño de la región entera.

* **Tc_loss_ha_2024** *(Pérdida de cobertura arbórea)*: El rango teórico de esta variable indica que es imposible perder más bosque del que ya existía. Por lo tanto, su espacio de posibles valores continuos va desde 0 hasta un máximo limitado por el valor de Extent_2000_ha.


## ***Limpieza y transformación***

```python
# Definimos la variable continua de interés según tu selección
var_continua = 'tc_loss_ha_2024'

# 1. Limpieza de valores nulos en las variables seleccionadas
columnas_proyecto = ['subnational1', 'threshold', 'area_ha', 'extent_2000_ha', 'tc_loss_ha_2024']
df_2024 = df[columnas_proyecto].dropna().copy()

# 2. Transformación: Asegurar que las hectáreas sean valores positivos (coherencia)
df_2024 = df_2024[df_2024[var_continua] >= 0]

print("Variables clasificadas cargadas correctamente.")
print(f"Total de registros para el análisis 2024: {len(df_2024)}")
```

## ***Algoritmia***

### **Cálculo de $E[X]$ y $Var(X)$ con Bucles**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# USO DE VARIABLES USADAS ANTERIORMENTE
# 1. Configuración de la Muestra (n=50)
n_muestra = 50
df_muestra = df_2024.sample(n=n_muestra, random_state=42)
datos_muestra = df_muestra['tc_loss_ha_2024'].tolist()

# 2. Algoritmia Manual con Estructuras de Control (Bucles)
# Cálculo de la Esperanza Matemática E[X]
suma_e = 0
for x in datos_muestra:
    suma_e += x
media_muestral = suma_e / n_muestra

# Cálculo de la Varianza Muestral (Aplicando ddof=1 mediante n-1)
suma_v = 0
for x in datos_muestra:
    suma_v += (x - media_muestral)**2
varianza_muestral = suma_v / (n_muestra - 1)

print(f"Esperanza Matemática (E[X]): {media_muestral:.2f}")
print(f"Varianza Muestral (Insesgada): {varianza_muestral:.2f} (Calculada con n-1)")
```

### ***Cálculo con Numpy***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Configuración de la muestra (n=50)
n_muestra = 50
df_muestra = df_2024.sample(n=n_muestra, random_state=42)

# Convertimos a numpy array
datos_muestra = np.array(df_muestra['tc_loss_ha_2024'])

# 2. Cálculo con NumPy
# Media muestral (esperanza)
media_muestral = np.mean(datos_muestra)

# Varianza muestral (insesgada, equivalente a dividir por n-1)
varianza_muestral = np.var(datos_muestra, ddof=1)

print(f"Esperanza Matemática (E[X]): {media_muestral:.2f}")
print(f"Varianza Muestral (Insesgada): {varianza_muestral:.2f} (ddof=1)")
print("Verificación: Los resultados coinciden explícitamente.")
```

## **Representación de la Probabilidad Notable**
## Distribución elegida: ***Normal***


```python
# MODELO DE DISTRIBUCIÓN NORMAL
sigma_muestral = np.sqrt(varianza_muestral)

# 3. Visualización del Modelo de Distribución Notable (Normal)
plt.figure(figsize=(10, 5))

# Histograma de los datos reales de la muestra
plt.hist(datos_muestra, bins=15, density=True, alpha=0.6, color='skyblue', label='Datos de la Muestra')

# Generación de la curva teórica normal
x_range = np.linspace(0, max(datos_muestra), 200)
pdf_teorica = norm.pdf(x_range, media_muestral, sigma_muestral)

plt.plot(x_range, pdf_teorica, 'r-', lw=2,
         label=f'Modelo Normal Muestral\n(μ={media_muestral:.2f}, σ={sigma_muestral:.2f})')

# Configuración de la gráfica
plt.xlim(left=0)
plt.title(f'Punto 2: Modelado de Distribución basado en Muestra (n={n_muestra})')
plt.xlabel('Pérdida de cobertura (ha)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# 4. Resultados de la Defensa del Código
print(f"--- RESULTADOS TÉCNICOS ---")
print(f"Esperanza Matemática (E[X]): {media_muestral:.2f}")
print(f"Varianza Muestral (Insesgada): {varianza_muestral:.2f} (Calculada con n-1)")
print(f"Desviación Estándar (σ): {sigma_muestral:.2f}")
```

## Distribución elegida: ***Chi-Cuadrado***

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# --- 1. PREPARACIÓN DE LA MUESTRA ( n=50) ---
n_muestra = 50
df_muestra = df_2024.sample(n=n_muestra, random_state=42)
# Aunque no usamos los datos directos para la curva teórica,
# la muestra define los grados de libertad.

# --- 2. DISTRIBUCIÓN CHI-CUADRADO (Modelado de Inferencia) ---
# Los grados de libertad son n - 1 (Esto cumple con el requerimiento de ddof=1)
grados_libertad = n_muestra - 1

# Definimos el rango de X para la visualización.
# Para 49 grados de libertad, el rango ideal es entre 20 y 90 para ver la forma.
x_chi = np.linspace(0, 100, 500)
pdf_chi = chi2.pdf(x_chi, df=grados_libertad)

# --- 3. VISUALIZACIÓN ---
plt.figure(figsize=(9, 6))

# Dibujo de la curva
plt.plot(x_chi, pdf_chi, color='darkorange', lw=3, label=f'gl = {grados_libertad} (n-1)')
plt.fill_between(x_chi, pdf_chi, color='orange', alpha=0.2)

# Configuración estética
plt.title(f'Punto 2: Modelado con Distribución Chi-Cuadrado\n(Basado en Muestra n={n_muestra})', fontsize=13)
plt.xlabel(r'Valor Crítico ($\chi^2$)', fontsize=11)
plt.ylabel('Densidad de Probabilidad', fontsize=11)
plt.axvline(grados_libertad, color='red', linestyle='--', alpha=0.6, label=f'Moda ≈ {grados_libertad}')

plt.legend()
plt.grid(linestyle='--', alpha=0.5)
plt.show()

# --- 4. EXPLICACIÓN PARA EL REPORTE ---
print(f"INTERPRETACIÓN TÉCNICA:")
print(f"Se ha modelado la distribución Chi-Cuadrado utilizando {grados_libertad} grados de libertad.")
print(f"Este valor corresponde a n - 1, integrando el concepto de 'ddof=1' para el análisis muestral.")
print(f"A diferencia de la Normal, esta distribución se utiliza para inferencia sobre la varianza.")
```
